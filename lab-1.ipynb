```python
import numpy as np
import matplotlib.pyplot as plt 

sf = 100
ts = 1.0/sf

datax = np.arange(0,2,ts)

datay1 = np.sin(2*np.pi*datax*10)
datay2 = np.sin(2*np.pi*datax*50)

fig, pic = plt.subplots(figsize=(8,4),layout="constrained")

# pic.plot(datax, datay2, label="X(nT)", color="red")
# pic.plot(datax, datay1, label="X(n)",color="green")

pic.grid()
pic.legend()
pic.set_xlabel("Time")
pic.set_ylabel("Amplitude")
pic.set_title("Signal Aliasing Effect")
plt.subplot(211)
plt.plot(datax, datay2, label="X(nT)", color="red")
plt.subplot(212)
plt.plot(datax, datay1, label="X(n)",color="green")
```

```python
import numpy as np
import matplotlib.pyplot as plt 

sf = 120
ts = 1.0/sf
t = np.arange(0,2,ts)

freq = 10
y = np.sin(2*np.pi*freq*t)

plt.figure(figsize=(8,10))
plt.subplot(211)
plt.plot(t,y,"blue")
plt.ylabel("Ampletude 1")

freq = 50
y = np.sin(2*np.pi*freq*t)
plt.figure(figsize=(8, 10))
plt.subplot(212)
plt.plot(t, y, "green")
plt.ylabel("Ampletude 2")
```
